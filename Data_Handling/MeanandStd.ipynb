```python
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
data = [-5, 1, 8, 7, 2]
a=np.mean(data)
print(a)
b=np.var(data)
c=math.sqrt(b)
print (c)
df = pd.DataFrame({
'data': data,
'mean': [a for i in range(len(data))],
'std': [c for i in range(len(data))]})
df.plot()
plt.show()
```
